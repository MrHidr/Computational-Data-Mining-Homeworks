## 1 Data prepration

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
X = pd.read_csv('secom.data', sep=' ', header=None)
y = pd.read_csv('secom_labels.data', sep=' ', header=None)
print("dataset shape: ", X.shape)
```

### 1.1 Data cleaning
- Removing columns with zero variance (constanc values)
- Median Imputation

```python
cons_columns = [col for col in X.columns if X[col].nunique() <= 1]
X_filtered = X.drop(columns=cons_columns)
print("removed columns with zero variance: ",len(cons_columns))
print("dataset shape after removing columns: ", X_filtered.shape)
```

```python
imputer = SimpleImputer(strategy='median')
X_imputed = pd.DataFrame(imputer.fit_transform(X_filtered), columns=X_filtered.columns)
print("Dataset columns filled with median imputation method")
```

### 1.2 Normalizing
while SVD is scale-sensetive so datas needed to be scaled to standard values.
scale-sensetive: Higher variance has Higher importancy in algorithm

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_scaled_array = scaler.fit_transform(X_imputed)
X_scaled = pd.DataFrame(X_scaled_array, columns=X_imputed.columns)
```

```python
print("mean before normalization: ",np.mean(X_imputed.iloc[:, 0]))
print("mean after normalization: ",np.mean(X_scaled.iloc[:, 0]))
print("--"*10)
print("variance before normalization: ",np.std(X_imputed.iloc[:, 0]))
print("variance after normalization: ",np.std(X_scaled.iloc[:, 0]))
```

After normalization: mean is near to 0 &
variance is near to 1
